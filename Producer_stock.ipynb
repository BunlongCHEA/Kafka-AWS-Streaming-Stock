```python
import pandas as pd
from kafka import KafkaProducer
from time import sleep
from json import dumps 
import json
```

```python
# Connect kafka producer to AWS server with encoder 'utf-8'
producer = KafkaProducer(bootstrap_servers=['43.206.231.106:9092'], #change ip here
                         value_serializer=lambda x: 
                         dumps(x).encode('utf-8'))
```

```python
# For test whether producer can send to consumer ot not
producer.send('stock_market_demo', value={'name':'Jeny'})
```

```python
# Now read csv from our directory using pandas
df = pd.read_csv("D:/ITStep/Apache_Kafka/stock-market-kafka-data-engineering-project-main/indexProcessed.csv")
df.head(500)
```

```python
# send message that read from pandas to consumer record by record, and sleep 1 second so not to overload server to consumer
while True:
    stocks = df.sample(1).to_dict(orient='records')[0]
    producer.send('stock_market_demo', value=stocks)
    sleep(1)
```

```python
# restart all the values in consumer
producer.flush()
```
